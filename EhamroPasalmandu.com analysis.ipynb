```python
fig = px.line(df, x='date', y='dailysales', title='EhamroPasalmandu.com Sales')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
df_rolling = df.set_index('date').rolling(365).mean().reset_index()
fig = px.line(df_rolling, x='date', y='dailysales', title='EhamroPasalmandu.com Rolling mean')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
df = df.set_index('date')
```

```python
df_month = df.resample('M').sum()
```

```python
fig = px.line(df_month, x=df_month.index, y='dailysales', title='EhamroPasalmandu.com Monthly mean')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
df_year = df.resample('Y').sum()
```

```python
fig = px.line(df_year, x=df_year.index, y='dailysales', title='EhamroPasalmandu.com Monthly mean')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
